```python
%matplotlib inline
url = "http://www.presidency.ucsb.edu/data/popularity.php"
import pandas as pd
from textblob import TextBlob
from nltk.corpus import stopwords
from collections import Counter
import string
import numpy as np
stops = stopwords.words('english') + list(string.punctuation)
stops.append("n't")
stops.append("ve")
stops.append("re")
```

```python
df = pd.read_html("data/approval_rating.txt")[0]
df = df[[0,1,2,4,5,6]]
df = df[3:]
df.columns = df.iloc[0]
df.drop(3, axis=0, inplace=True)
df.drop('President', axis=1, inplace=True)
df[['Start Date','End Date']] = df[['Start Date','End Date']].apply(lambda x: pd.to_datetime(x))
df.Approving = df.Approving.astype(int)
df.Disapproving = df.Disapproving.astype(int)
df['unsure/no data'] = df['unsure/no data'].astype(int)
```

```python
# src: http://www.presidency.ucsb.edu/sou.php
data = []
files = []
with open('data/sous_list.txt') as f:
    for line in f.readlines():
        data.append(line.strip(".txt\n").split('-')[2])
        files.append("data/" + line.strip("\n"))
data, files
```

```python
sotus = {}
for fi, date in zip(files, data):
    with open(fi) as f:
        sotus[date] = TextBlob(''.join(f.readlines()).decode('utf-8', 'ignore'))
```

```python
def countered(def_dict, num):
    return Counter(dict([(x,y) for x, y in def_dict.items() if x not in stops])).most_common(num)
```

```python
all_words = []
all_words2 = []
o_years = []
o_words = []
o_counts = []
for k, v in sotus.items():
    print k
    print countered(v.np_counts, 12)
    temp = countered(v.word_counts, 10)
    all_words += TextBlob(' '.join([word for word, c in temp])).words.singularize()
    all_words2 += [word for word, c in temp]
#     print Counter([tuple(val.lower()) for val in v.ngrams(3)]).most_common(10)
    print

all_words = list(set(all_words))
all_words.remove('u')
all_words.append("us")
all_words.remove('person')
all_words.append("people")
# print all_words2
for k, v in sotus.items():
    print k
    for word in set(all_words):
        o_years.append(k)
        o_words.append(word)
        o_counts.append(v.word_counts[word])
#         out.append("%s,%s,%d" % (k, word, v.word_counts[word]))
        print word, v.word_counts[word]
#     print Counter([tuple(val.lower()) for val in v.ngrams(3)]).most_common(10)
    print
```
