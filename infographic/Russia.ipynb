#Interest Rates

```python
intrst = pd.read_html('data/russia/other_interest_rates.txt')[0]
intrst.columns = intrst.iloc[0]
intrst.drop(0,axis=0, inplace=True)
```

```python
intrst['percentage'] = intrst.percentage.apply(lambda x: float(x.split("%")[0]))
```

```python
intrst
```

```python
intrst[['region','percentage']].set_index('region').sort('percentage').to_clipboard()
```

```python
intrst_change = pd.read_html('data/russia/interest_rate_changes.txt')[0]
intrst_change.columns = intrst_change.iloc[0]
intrst_change.drop(0,axis=0, inplace=True)
intrst_change['percentage'] = intrst_change.percentage.apply(lambda x: float(x.split("%")[0]))
```

```python
intrst_change['date'] = intrst_change['change date'].apply(lambda x: pd.to_datetime(x))
intrst_change.drop('change date', axis=1,inplace=True)
```

```python
intrst_change['date2'] = intrst_change.date.apply(lambda x: x.strftime("%Y") + "-" + x.strftime("%B")[:3])
```

```python
intrst_change.sort(columns='date',inplace=True)
```

```python
intrst_change[['date2','percentage']].set_index('date2')
```

```python
intrst_change[['date2','percentage']].set_index('date2').to_clipboard()
```

```python
intrst_change
```

#Inflation Rates

```python
inf2 = pd.read_html('data/russia/other_inflation_rates.txt')[0]
inf2.columns = inf2.iloc[0]
inf2.drop(0,axis=0, inplace=True)
inf2['inflation'] = inf2.inflation.apply(lambda x: float(x.split("%")[0]))
inf2['period'] = inf2.period.apply(lambda x: pd.to_datetime(x))
inf2['period2'] = inf2.period.apply(lambda x: x.strftime("%Y") + "-" + x.strftime("%B")[:3])
inf2.sort(columns='period',inplace=True)
inf2.head()
```

```python
inf2[['period2','inflation']].set_index('period2').to_clipboard()
```

```python
inf = pd.read_html('data/russia/inflation_rate_changes.txt')[0]
inf.columns = inf.iloc[0]
inf.drop(0,axis=0, inplace=True)
inf['inflation'] = inf.inflation.apply(lambda x: float(x.split("%")[0]))
inf.head()
```

```python
inf['period'] = inf.period.apply(lambda x: pd.to_datetime(x))
```

```python
inf['period2'] = inf.period.apply(lambda x: x.strftime("%Y") + "-" + x.strftime("%B")[:3])
```

#Present Economic Situations

```python
sit = pd.read_csv('data/russia/present.csv',header=None, usecols=[0,1,2,3,4,5,6])
sit = sit.T.drop([7,8,9],axis=1)
sit.head()
```

```python
sit.columns = sit.iloc[0]
sit.drop(0,axis=0, inplace=True)
sit.columns = ["date"] + list(sit.columns[1:])
sit['date'] = sit.date.apply(pd.to_datetime)
sit['date'] = sit.date.apply(lambda x: pd.to_datetime(x.strftime("%m") + "-28-2014"))
```

```python
sit.head()
```

```python
sit['date2'] = sit.date.apply(lambda x: x.strftime("%Y") + "-" + x.strftime("%B")[:3])
```

```python
sit.head()
```

```python
sit.sort(columns='date',inplace=True)
```

```python
sit.drop('date', axis=1).set_index('date2').to_clipboard()
```

#What Will Happen

```python
fut = pd.read_csv('data/russia/econ.csv',header=None)
fut = fut.T
fut.columns = fut.iloc[0]
fut.drop(0,axis=0, inplace=True)
fut.head()
```

```python
fut.columns = ["date"] + list(fut.columns[1:])
fut['date'] = fut.date.apply(pd.to_datetime)
fut.dropna(inplace=True)
```

```python
fut['date2'] = fut.date.apply(lambda x: x.strftime("%Y") + "-" + x.strftime("%B")[:3])
fut.sort(columns='date',inplace=True)
```

```python
fut.drop('date', axis=1).set_index('date2').to_clipboard()
```
